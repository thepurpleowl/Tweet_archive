```python
%matplotlib inline
import xlrd 
import numpy as np
wb=xlrd.open_workbook('tweets.xlsx')
for sheet in wb.sheets():
    r=sheet.nrows
    #print r
    tweets=[]
    labels=[]
    for i in range(1,639):
        l=sheet.row(i)
        l=str(l[0])
        l=l.replace("text:u","")
        l=l[1:-1]
        l=l.split(" ")
        #print l
        temp=[0,0,0,0,0]
        for j in range(1,len(l)):
            if l[j].startswith("@"):
                temp[0]=1
            if l[j].startswith("#"):
                temp[1]=1
            if l[j].startswith(("www","http","https")):
                temp[2]=1
            if l[j].find('@')==1:
                temp[3]=1
            else:
                temp[4]=1
        l=np.array(l)
        tweets.append(temp)
    for i in range(1,501):
        labels.append(1)
    for i in range(501,639):
        labels.append(0)
# print tweets
# print labels
# for i in range(1,len(tweets)):
#     print tweets[i], ":",labels[i]
print len(tweets)
print len(labels)
from sklearn.cross_validation import train_test_split 

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
clf=SVC()

from sklearn import tree
decisionTree=tree.DecisionTreeClassifier()

from sklearn.neighbors import KNeighborsClassifier
decisionKN=KNeighborsClassifier()

iter_list=np.arange(0.1,1,0.1)
svm_acc=[]
dt_acc=[]
knn_acc=[]
print iter_list
for i in iter_list:
    X_train,X_test, y_train, y_test=train_test_split(tweets,labels, test_size=i)

    clf.fit(X_train,y_train)
    pred=clf.predict(X_test)
    svm_acc.append(accuracy_score(y_test,pred))

    decisionTree.fit(X_train,y_train)
    treeprediction=decisionTree.predict(X_test)
    dt_acc.append(accuracy_score(y_test,treeprediction))

    decisionKN.fit(X_train,y_train)
    KNprediction=decisionKN.predict(X_test)
    knn_acc.append(accuracy_score(y_test,KNprediction))

import matplotlib.pyplot as plt
l1=plt.plot(iter_list,svm_acc,label="SVM")
l2=plt.plot(iter_list,dt_acc,label="Decision Tree",marker="o",linestyle="--")
l3=plt.plot(iter_list,knn_acc,label="KNN",linestyle="dashdot",marker="*")
plt.title("Accuracy Comparison with different models",fontweight="bold")
plt.xlabel("Train-Test Split",fontweight="bold")
plt.ylabel("Accuracy Score",fontweight="bold")
plt.legend(loc=0,fontsize=10)
plt.savefig('mod_acc.png')
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://www.jagran.com/"

r  = requests.get(url)
data = r.text
print r.status_code
soup = BeautifulSoup(data)

for link in soup.findAll('span',{'class':'tabtxt'}):
    print(link.getText()),'\n'
```

```python
url = "http://www.bhaskar.com/"

r  = requests.get(url)
data = r.text
print r.status_code
soup = BeautifulSoup(data)

for link in soup.findAll('a',{'class':'flicker'}):
    print link.get('title'),'\n'
```
